**Line graph:** Line graph is used to detect the movement of vehicle in college campus .In which the Blue color shows the vehicles entering the college campus and Red color shows the vehicles moving out of the college campus with respect to time.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Path to your Excel file
file_path = 'Intel_vehicle.xlsx'

# Load the Excel file
excel_data = pd.read_excel(file_path, sheet_name='Sheet1')

# Ploting the size
plt.figure(figsize=(10, 10))

#labelling
plt.plot([],[],color='blue',label='time_in',linewidth=5)
plt.plot([],[],color='red',label='time_out',linewidth=5)

# ploting the line graph
plt.plot(excel_data['Time_in'], marker='o', color='blue', linestyle='-')
plt.plot(excel_data['Time_out'], marker='o', color='red', linestyle='-')

plt.title('Vehicle in/out information with respect to time')
plt.xlabel('Time')
plt.ylabel('The vehicles entered in the campus')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Area graph:**  Area graph shows the parking occupancy in the campus.The lightcoral color shows the total area present for parking in the campus and lightblue color shows the area available for parking with respect time.

```python
import matplotlib.pyplot as plt

# Path to your Excel file
file_path = 'Intel_vehicle.xlsx'

# Load the Excel file
excel_data = pd.read_excel(file_path, sheet_name='Sheet1')

#ploting the size
plt.figure(figsize=(10,10))

#labelling
plt.plot([],[],color='lightblue',label='occupied area',linewidth=5)

#Total area available for parking
plt.axhspan(0,10,color='lightcoral')

#area occupied by parking
plt.stackplot(excel_data['Time_in'],excel_data['Si No.'],color='lightblue')

plt.title('parrking occupancy')
plt.xlabel('Time')
plt.ylabel('Availability of area for parking')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

**Piechart:**Piechart shows the authorized and unauthorized vehicles entering the college campus.Green color shows the percentage of vehicles which are authorized and Red color shows the percentage of vechiles which are unauthorized.

```python
import matplotlib.pyplot as plt

# Path to your Excel file
file_path = 'Intel_vehicle.xlsx'

# Load the Excel file
excel_data = pd.read_excel(file_path, sheet_name='Sheet1')

# Count the number of authorized and unauthorized entries
authorized_count = excel_data['Authorized_status'].sum()
unauthorized_count = len(excel_data) - authorized_count

# Data for the pie chart
labels = 'Authorized', 'Unauthorized'
sizes = [authorized_count, unauthorized_count]
colors = ['green', 'red']
explode = (0.1, 0)  # explode the 1st slice

# Plotting the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Authorized vs Unauthorized Vehicles')
plt.show()
```
